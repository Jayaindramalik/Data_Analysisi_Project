```python
df['Kode Barang'].value_counts()
```

```python
ax=df['Kode Barang'].value_counts().plot(kind='pie',
    figsize=(10,6),
        autopct='%1.1f%%')
```

```python
ax=df['Kode Barang'].value_counts().plot(kind='bar',figsize=(10,6))
ax.set_ylabel('Total Pembelian jenis Item')
for container in ax.containers:
    ax.bar_label(container)
```

```python
df.groupby('Kode Barang')['total'].sum()
```

```python
ax=df.groupby('Kode Barang')['total'].sum()
ax=ax.sort_values(ascending=False)
ax.plot(kind='bar',figsize=(10,6))
```
